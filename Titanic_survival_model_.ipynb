```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the data
titanic_data = pd.read_csv("/content/tested.csv")
```

```python
titanic_data.head()
```

```python
titanic_data.shape
```

```python
titanic_data.info()
```

```python
#sum of null values in dataset
titanic_data.isnull().sum()
```

```python
#handling missing values by dropping the cabin column
titanic_data = titanic_data.drop(columns='Cabin', axis=1)
```

```python
#handling the missing values by taking mean in the age column
titanic_data["Age"].fillna(titanic_data["Age"].mean(),inplace=True)
```

```python
#handling the missing value by taking the mode in the fare column
titanic_data["Fare"].fillna(titanic_data["Fare"].mode()[0],inplace=True)
```

```python
#checking the missing values again to see the applied changes
titanic_data.isnull().sum()
```

```python
#statistical measures from dataset
titanic_data.describe()
```

```python
# number of people who survived and number of people who didn't survived
titanic_data['Survived'].value_counts()
```

```python
#visualization of survived column
sns.countplot(x='Survived', data=titanic_data)
```

```python
sns.countplot(x='Sex', data=titanic_data)
plt.show()
```

```python
titanic_data['Sex'].value_counts()
```

```python
titanic_data["Embarked"].value_counts()
```

```python
# converting categorical Columns

titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

```python
#seperating features and target
#features - passengerID, Name, Ticket, Survived
#target - survived
X = titanic_data.drop(columns=["PassengerId","Name","Ticket","Survived"],axis=11)
Y = titanic_data["Survived"]
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#model training - linear regression
model = LogisticRegression()
```

```python
# training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

```python
#checking the accuracy score
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
