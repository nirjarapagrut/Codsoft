```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading dataset
credit_card= pd.read_csv("/content/creditcard.csv")
```

```python
credit_card.head(5)
```

```python
credit_card.tail()
```

```python
credit_card.info()
```

```python
#sum of all missing values
credit_card.isnull().sum()
```

```python
#counting total class values
credit_card['Class'].value_counts()
```

```python
# separating the data
#0 - normal trasaction
#1 - fraud transaction
#0.004628
normal = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
```

```python
print(normal.shape)
print(fraud.shape)
```

```python
#describing normal transactions from the data
normal.Amount.describe()
```

```python
#describing fraud transactions from the data
fraud.Amount.describe()
```

```python
# comparing normal transactions and fraud transactions
credit_card.groupby('Class').mean()
```

```python
#creating a sample data containing similar distribution of fruad transactions and normal transaction
normal_sample = normal.sample(n=492)
```

```python
#concating the new data with old dataset
new_credit_card = pd.concat([normal_sample, fraud],axis = 0)
```

```python
new_credit_card.head()
```

```python
new_credit_card.head(5)
```

```python
new_credit_card.tail(5)
```

```python
# now we have same amount of fraud transaction as normal transaction
new_credit_card['Class'].value_counts()
```

```python
#taking mean of both type of transaction
new_credit_card.groupby('Class').mean()
```

```python
#splitting the data
X = new_credit_card.drop(columns='Class', axis=1)
Y = new_credit_card['Class']
```

```python
X.head(5)
```

```python
Y.head(5)
```

```python
#splitting data and testing the training data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#creating a logistic regression model
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
#checking accuracy of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
#checking accuracy of testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```
